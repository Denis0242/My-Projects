# **Sets - Access**

### Definition: Sets in Python are unordered collections of unique elements. You cannot access items in a set by referring to an index, since sets are unordered.




### Example 1: Accessing elements using a loop


```python
fruits = {"apple", "banana", "cherry"}
print(type(fruits))

for fruit in fruits:
    print(fruit)

# Note: The order of items is not guaranteed.
```

```python
for i in fruits:
  print(i)
```

```python
fruits ={'apple','banana','cherry'}
my_list = list(fruits)
print(my_list)
print(my_list[0])  # Now you can ac"cess by index
```

### Example 2: Checking if an item exists in a set


```python
fruits ={'apple','banana','cherry'}
print("apple" in fruits)
```

### Example 3: Accessing elements using set operations
* Use pop() to Access and Remove an Arbitrary Element

```python
fruits = {"apple", "banana", "cherry"}
print(fruits.pop())  # Removes and returns an arbitrary element
print(fruits)
```

# **Sets - Add**

### Definition: Adding elements to a set is done using the `add()` method for single elements, or the `update()` method for adding multiple elements.



### Example 1: Adding a single item with add()


```python
fruits = {"apple", "banana", "cherry"}
fruits.add("orange")
print(fruits)
```

### Example 2: Adding multiple items with update()


```python
fruits = {"apple", "banana", "cherry"}
fruits.update(["mango", "grape"])
print(fruits)
```

### Example 3: Adding items from another set


```python
fruits = {"apple", "banana", "cherry"}
tropical = {"pineapple", "papaya"}
fruits.update(tropical)
print(fruits)
```

```python
# Creating a set
set1 = {1, 2, 3}

# Add one item
set1.add(4)

# Add multiple items
set1.update([5, 6])

print(set1)
```

# **Sets - Remove**

### Definition: Removing elements from a set can be done using `remove()` or `discard()`. The `remove()` method will raise an error if the item does not exist, while `discard()` will not.



```python
# Using Remove Method
set1 = {1, 2, 3, 4, 5}
set1.remove(3)
print('operation 1 :',set1)

set1 = {1, 2, 3, 4, 5}
# Using discard() Method
set1.discard(4)
print('operation 2: ',set1)

set1 = {1, 2, 3, 4, 5}
val = set1.pop() # some random element from set will be removed
print('pop value :',val)
print('elements after pop method:',set1)
```

### Example 1: Removing an item with remove()


```python
fruits = {"apple", "banana", "cherry"}
fruits.remove("banana")
print(fruits)
```

### Example 2: Removing an item with discard()


```python
fruits = {"apple", "banana", "cherry"}
fruits.discard("apple")
print(fruits)
```

### Example 3: Removing an arbitrary item with pop()


```python
fruits = {"apple", "banana", "cherry"}
popped_item = fruits.pop()
print(popped_item)
print(fruits)
```

# **Sets - Loop**

### Definition: Looping through a set involves iterating over each element in the set using a loop.



### Example 1: Looping through a set with a for loop


```python
fruits = {"apple", "banana", "cherry"}
for fruit in fruits:
    print(fruit)
```

### Example 2: Looping through a set and modifying elements


```python
fruits = {"apple", "banana", "cherry"}

for fruit in fruits.copy():
    fruits.add(fruit + " juice")
print(fruits)
```

### Example 3: Using a while loop to iterate through the set


```python
fruits = {"apple", "banana", "cherry"}
while fruits:
    print(fruits.pop())

print(fruits)
```

# **Sets - Join**

### Definition: Joining sets involves combining two or more sets using set operations like union, intersection, difference, and symmetric difference.



| Operation      | Symbol/Method            | Example                        |                           |
| -------------- | ------------------------ | ------------------------------ | ------------------------- |
| Union          | \`                       | `or`.union()\`                 | Combines all unique items |
| Intersection   | `&` or `.intersection()` | Common items only              |                           |
| Difference     | `-` or `.difference()`   | Items in one but not the other |                           |
| Symmetric Diff | `^`                      | Items in either but not both   |                           |


### Example 1: Union of two sets


```python
set1 = {"apple", "banana", "cherry","mango"}
set2 = {"pineapple", "mango", "papaya"}
union_set = set1.union(set2)
print(union_set)
```

### Example 2: Intersection of two sets


```python
set1 = {"apple", "banana", "cherry","mango"}
set2 = {"pineapple", "mango", "papaya"}
intersection_set = set1.intersection(set2)
print(intersection_set)
```

### Example 3: Symmetric difference between two sets


```python
set1 = {"apple", "banana", "cherry","mango","orange"}
set2 = {"pineapple", "mango", "papaya","orange"}
sym_diff_set = set1.symmetric_difference(set2)
print(sym_diff_set)
```

# **Sets - Set Methods**

### Definition: Set methods are built-in functions that perform specific operations on set elements.



### Example 1: Using add() to add an element to a set


```python
fruits = {"apple", "banana", "cherry"}
fruits.add("orange")
print(fruits)
```

### Example 2: Using clear() to empty a set


```python
fruits = {"apple", "banana", "cherry"}
fruits.clear()
print(fruits)
```

### Example 3: Using copy() to create a shallow copy of a set


```python
fruits = {"apple", "banana", "cherry"}
fruits_copy = fruits.copy()
print(fruits_copy)
```
